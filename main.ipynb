```python
from dotenv import load_dotenv
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq

load_dotenv()
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
question_prompt = ChatPromptTemplate.from_template("""
You're a Java technical interviewer 
Generate One clear interview question about this topic : {topic}
Just the question, nothing else
 """)

eval_prompt = ChatPromptTemplate.from_template(""" 
You are a Java technical interviewer eveluating an answer

Question : {question}
Canditate's Answer: {answer}
Give:
1. Correct/Incorrect 
2. Brief feedback (2-3 sentences)
3. The ideal answer in simple terms
""")
```

```python
question_chain = question_prompt | llm
eval_chain = eval_prompt | llm
```

```python
topic = "Spring Boot dependency injection"

question = question_chain.invoke({"topic": topic})
print(f"📌 Question:\n{question.content}")
```

```python
answer = "I use @Autowired to automatically inject a bean by type. If there are multiple beans of the same type, I use @Qualifier to specify which one to inject by name."

feedback = eval_chain.invoke({
    "question": question.content,
    "answer": answer
})

print(f"💬 Feedback:\n{feedback.content}")
```

```python
def run_interview():
    print("🎯 Java Interview Coach\n")
    topic = input("Enter a topic (e.g. OOP, Collections, JVM, Spring Boot): ")

    question = question_chain.invoke({"topic": topic})
    print(f"\n📌 Question:\n{question.content}\n")

    answer = input("Your answer: ")

    feedback = eval_chain.invoke({
        "question": question.content,
        "answer": answer
    })
    print(f"\n💬 Feedback:\n{feedback.content}")

run_interview()
```
